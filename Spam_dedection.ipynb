```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
class ANNModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1=nn.Linear(in_features=len(vectorizer.vocabulary_),out_features=256)
        self.relu=nn.ReLU()
        self.l2=nn.Linear(in_features=256, out_features=128)
        self.tanh=nn.Tanh()
        self.l3=nn.Linear(in_features=128,out_features=2)
    def forward(self,x):
        x=x.view(-1,len(vectorizer.vocabulary_))
        x=self.relu(self.l1(x))
        x=self.tanh(self.l2(x))
        return self.l3(x)
```

```python
class SpamTrainDataset(Dataset):
    def __init__(self,emails,labels,vectorizer):
        super().__init__()
        self.emails=emails
        self.labels=labels
    def __len__(self):
        return(len(self.emails))
    def __getitem__(self,idx):
        feature=self.emails[idx]
        label=self.labels[idx]
        
        feature=vectorizer.transform([feature]).todense()
        label=1 if label=="spam" else 0
        return {"feature":torch.tensor(feature, dtype=torch.float32),"label":torch.tensor(label, dtype=torch.long)}
```

```python
class SpamTestDataset(Dataset):
    def __init__(self,emails,vectorizer):
        super().__init__()
        self.emails=emails
    def __len__(self):
        return(len(self.emails))
    def __getitem__(self,idx):
        feature=self.emails[idx]
        
        feature=vectorizer.transform([feature]).todense() 
    
        return {"feature":torch.tensor(feature, dtype=torch.float32)}
```

```python
train_dataset=SpamTrainDataset(train_feature,train_label,vectorizer)
```

```python
test_dataset=SpamTestDataset(test_feature,vectorizer)
```

```python
train_dataloader=DataLoader(dataset=train_dataset,batch_size=64, shuffle=True)
```

```python
test_dataloader=DataLoader(dataset=test_dataset,batch_size=64, shuffle=False)
```

```python
ann=ANNModel()
```

```python
loss_fn=nn.CrossEntropyLoss()
```

```python
optimizer=torch.optim.SGD(params=ann.parameters(),lr=1e-2)
```

```python
for epoch in range(20):
    total_loss=0
    for batch in train_dataloader:
        features=batch["feature"]
        labels=batch["label"]
        y_pred=ann(features).squeeze()
        loss=loss_fn(y_pred,labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss+=loss.item()
    print(total_loss/len(train_dataloader))
```

```python
ann=ann.eval()
all_predicted=[]
with torch.no_grad():
    for data in test_dataloader:
        features=data["feature"]
        predicted=ann(features)
        predicted=torch.sigmoid(predicted).detach().numpy()
        predicted=predicted.argmax(axis=1)
        all_predicted.extend(predicted.tolist())
```
